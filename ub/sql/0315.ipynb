## 해보기
1. 그룹화 국가
2. 출력 -> country, country 갯수를 구하는 dataframe 만들기
3. country 갯수를 기준으로 내림차순 정렬

```python
sql_3 = '''
        SELECT
        Country,
        COUNT(`Country`) AS 'cnt'
        FROM sales
        GROUP BY Country
        ORDER BY `cnt` DESC
        '''
result_3 = _db.executeAll(sql_3)
df_3 = pd.DataFrame(result_3)
df_3
```
